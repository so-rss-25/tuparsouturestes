# Import dataset brut

```python
import pandas as pd
```

```python
url = f"https://docs.google.com/spreadsheets/d/1gt391glXJIkt1zwGs-kC3EkL7J1R4BHafvyiIifcYH8/gviz/tq?tqx=out:csv&sheet=dataset_brut_v0"
df = pd.read_csv(url)
df
```

# Sélection des colonnes nécessaires à l'analyse

```python
df=df.drop(columns=["DailyRate","EmployeeCount","HourlyRate","MonthlyRate","Over18","StandardHours"])
df.info()
```

# Traduction des titres de colonnes

```python
df.columns=["age","attrition","deplacements_pro","service","trajet_quotidien_en_miles","num_niveau_detude","domaine_detude","identifiant_employe","num_satisfaction_environnement","genre","num_implication","num_niveau_hierarchique","poste","num_satisfaction_travail","situation","salaire_mensuel","num_nb_entreprises","heures_sup","num_pourcentage_augmentation","num_performance","num_satisfaction_relation","num_taux_daction","tt_annee_travail","num_temps_de_formation","num_equilibre_pro_perso","annees_dans_entreprise","annees_poste_actuel","annees_depuis_promotion","num_annees_avec_manager"]
df
```

# Traduction des colonnes

```python
# attrition
df.loc[df["attrition"] =="Yes","attrition"] = "Oui"
df.loc[df["attrition"] =="No","attrition"] = "Non"

# deplacements_pro
df.loc[df["deplacements_pro"] =="Travel_Rarely","deplacements_pro"] = "Rare"
df.loc[df["deplacements_pro"] =="Travel_Frequently","deplacements_pro"] = "Fréquent"
df.loc[df["deplacements_pro"] =="Non-Travel","deplacements_pro"] = "Aucun"

# service
df.loc[df["service"] =="Sales","service"] = "Ventes"
df.loc[df["service"] =="Research & Development","service"] = "R&D"
df.loc[df["service"] =="Human Resources","service"] = "RH"

# domaine_detude
df.loc[df["domaine_detude"] =="Life Sciences","domaine_detude"] = "Sciences"
df.loc[df["domaine_detude"] =="Other","domaine_detude"] = "Autre"
df.loc[df["domaine_detude"] =="Medical","domaine_detude"] = "Médecine"
df.loc[df["domaine_detude"] =="Human Resources","domaine_detude"] = "RH"
df.loc[df["domaine_detude"] =="Technical Degree","domaine_detude"] = "Techonologies"

# genre
df.loc[df["genre"] =="Female","genre"] = "Femme"
df.loc[df["genre"] =="Male","genre"] = "Homme"

# poste
df.loc[df["poste"] =="Healthcare Representative","poste"] = "Délégué médical"
df.loc[df["poste"] =="Human Resources","poste"] = "RH"
df.loc[df["poste"] =="Laboratory Technician","poste"] = "Technicien de laboratoire"
df.loc[df["poste"] =="Research Scientist","poste"] = "Chercheur"
df.loc[df["poste"] =="Manufacturing Director","poste"] = "Directeur de production"
df.loc[df["poste"] =="Sales Representative","poste"] = "Commercial"
df.loc[df["poste"] =="Research Director","poste"] = "Directeur R&D"
df.loc[df["poste"] =="Sales Executive","poste"] = "Responsable commercial"

# situation
df.loc[df["situation"] =="Single","situation"] = "Célibataire"
df.loc[df["situation"] =="Married","situation"] = "Marié(e)"
df.loc[df["situation"] =="Divorced","situation"] = "Divorcé(e)"

# heures_sup
df.loc[df["heures_sup"] =="Yes","heures_sup"] = "Oui"
df.loc[df["heures_sup"] =="No","heures_sup"] = "Non"

df
```

# Correspondance des colonnes

```python
# attrition
df.loc[df["attrition"] =="Oui","num_attrition"] = 1
df.loc[df["attrition"] =="Non","num_attrition"] = 0

# niveau d'étude
df.loc[df["num_niveau_detude"] ==1,"niveau_detude"] = "Baccalauréat"
df.loc[df["num_niveau_detude"] ==2,"niveau_detude"] = "Bac+2"
df.loc[df["num_niveau_detude"] ==3,"niveau_detude"] = "Licence"
df.loc[df["num_niveau_detude"] ==4,"niveau_detude"] = "Master"
df.loc[df["num_niveau_detude"] ==5,"niveau_detude"] = "Doctorat"

# satisfaction de l'environnement
df.loc[df["num_satisfaction_environnement"] ==1,"satisfaction_environnement"] = "Faible"
df.loc[df["num_satisfaction_environnement"] ==2,"satisfaction_environnement"] = "Moyenne"
df.loc[df["num_satisfaction_environnement"] ==3,"satisfaction_environnement"] = "Élevée"
df.loc[df["num_satisfaction_environnement"] ==4,"satisfaction_environnement"] = "Très élevée"

# niveau d'implication
df.loc[df["num_implication"] ==1,"implication"] = "Faible"
df.loc[df["num_implication"] ==2,"implication"] = "Moyenne"
df.loc[df["num_implication"] ==3,"implication"] = "Élevée"
df.loc[df["num_implication"] ==4,"implication"] = "Très élevée"

# niveau hiérarchique
df.loc[df["num_niveau_hierarchique"] ==1,"niveau_hierarchique"] = "Débutant"
df.loc[df["num_niveau_hierarchique"] ==2,"niveau_hierarchique"] = "Confirmé"
df.loc[df["num_niveau_hierarchique"] ==3,"niveau_hierarchique"] = "Senior"
df.loc[df["num_niveau_hierarchique"] ==4,"niveau_hierarchique"] = "Manager"
df.loc[df["num_niveau_hierarchique"] ==5,"niveau_hierarchique"] = "Directeur"

# satisfaction du travail
df.loc[df["num_satisfaction_travail"] ==1,"satisfaction_travail"] = "Faible"
df.loc[df["num_satisfaction_travail"] ==2,"satisfaction_travail"] = "Moyenne"
df.loc[df["num_satisfaction_travail"] ==3,"satisfaction_travail"] = "Élevée"
df.loc[df["num_satisfaction_travail"] ==4,"satisfaction_travail"] = "Très élevée"

# performance
df.loc[df["num_performance"] ==1,"performance"] = "Faible"
df.loc[df["num_performance"] ==2,"performance"] = "Moyenne"
df.loc[df["num_performance"] ==3,"performance"] = "Élevée"
df.loc[df["num_performance"] ==4,"performance"] = "Très élevée"

# satisfaction des relations au travail
df.loc[df["num_satisfaction_relation"] ==1,"satisfaction_relation"] = "Faible"
df.loc[df["num_satisfaction_relation"] ==2,"satisfaction_relation"] = "Moyenne"
df.loc[df["num_satisfaction_relation"] ==3,"satisfaction_relation"] = "Élevée"
df.loc[df["num_satisfaction_relation"] ==4,"satisfaction_relation"] = "Très élevée"

# taux d'action
df.loc[df["num_taux_daction"] ==0,"taux_daction"] = "Aucun"
df.loc[df["num_taux_daction"] ==1,"taux_daction"] = "Faible"
df.loc[df["num_taux_daction"] ==2,"taux_daction"] = "Moyen"
df.loc[df["num_taux_daction"] ==3,"taux_daction"] = "Haut"

# équilibre vie pro / vie perso
df.loc[df["num_equilibre_pro_perso"] ==1,"equilibre_pro_perso"] = "Mauvais"
df.loc[df["num_equilibre_pro_perso"] ==2,"equilibre_pro_perso"] = "Moyen"
df.loc[df["num_equilibre_pro_perso"] ==3,"equilibre_pro_perso"] = "Bon"
df.loc[df["num_equilibre_pro_perso"] ==4,"equilibre_pro_perso"] = "Très bon"

df
```

# Segmentation des données

```python
# segmentation de l'âge
df["categorie_age"] = "55 et plus"
df.loc[df["age"] <=54,"categorie_age"] = "45 - 54"
df.loc[df["age"] <=44,"categorie_age"] = "35 - 44"
df.loc[df["age"] <=34,"categorie_age"] = "25 - 34"
df.loc[df["age"] <=24,"categorie_age"] = "24 et moins"

# segmentation du trajet_quotidien_en_miles
df["categorie_trajet"] = "21+"
df.loc[df["trajet_quotidien_en_miles"] <=20,"categorie_trajet"] = "11 - 20"
df.loc[df["trajet_quotidien_en_miles"] <=10,"categorie_trajet"] = "06 - 10"
df.loc[df["trajet_quotidien_en_miles"] <=5,"categorie_trajet"] = "0 - 05"

# segmentation des catégories de postes
df.loc[df["poste"] == "Chercheur","statut"] = 1
df.loc[df["poste"] == "Technicien de laboratoire","statut"] = 1
df.loc[df["poste"] == "Commercial","statut"] = 1
df.loc[df["poste"] == "RH","statut"] = 1
df.loc[df["poste"] == "Responsable commercial","statut"] = 2
df.loc[df["poste"] == "Directeur de production","statut"] = 2
df.loc[df["poste"] == "Délégué médical","statut"] = 2
df.loc[df["poste"] == "Manager","statut"] = 3
df.loc[df["poste"] == "Directeur R&D","statut"] = 3

# segmentation du salaire
df["categorie_salaire"] = "9861 +"
df.loc[df["salaire_mensuel"] <=9860,"categorie_salaire"] = "5744 - 9860"
df.loc[df["salaire_mensuel"] <=5743,"categorie_salaire"] = "4230 - 5743"
df.loc[df["salaire_mensuel"] <=4229,"categorie_salaire"] = "2697 - 4229"
df.loc[df["salaire_mensuel"] <=2696,"categorie_salaire"] = "0 - 2696"

# segmentation du nombre d'entreprises
df["nb_entreprises"] = "6 +"
df.loc[df["num_nb_entreprises"] <=5,"nb_entreprises"] = "4 - 5"
df.loc[df["num_nb_entreprises"] <=2,"nb_entreprises"] = "2 - 3"
df.loc[df["num_nb_entreprises"] <=1,"nb_entreprises"] = "0 - 1"

# segmentation du pourcentage d'augmentation
df["pourcentage_augmentation"] = "21 +"
df.loc[df["num_pourcentage_augmentation"] <=20,"pourcentage_augmentation"] = "15 - 20"
df.loc[df["num_pourcentage_augmentation"] <=14,"pourcentage_augmentation"] = "13 - 14"
df.loc[df["num_pourcentage_augmentation"] <=12,"pourcentage_augmentation"] = "11 - 12"

# segmentation du nombre d'années travaillées
df["niveau_carriere"] = "21 +"
df.loc[df["tt_annee_travail"] <=20,"niveau_carriere"] = "10 - 20"
df.loc[df["tt_annee_travail"] <=10,"niveau_carriere"] = "06 - 10"
df.loc[df["tt_annee_travail"] <=5,"niveau_carriere"] = "03 - 05"
df.loc[df["tt_annee_travail"] <=2,"niveau_carriere"] = "0 - 02"

# segmentation du temps de formation
df["temps_de_formation"] = "04 +"
df.loc[df["num_temps_de_formation"] <=3,"temps_de_formation"] = "01 - 03"
df.loc[df["num_temps_de_formation"] ==0,"temps_de_formation"] = "0"

# segmentation de l'ancienneté
df["anciennete"] = "21 +"
df.loc[df["annees_dans_entreprise"] <=20,"anciennete"] = "10 - 20"
df.loc[df["annees_dans_entreprise"] <=10,"anciennete"] = "06 - 10"
df.loc[df["annees_dans_entreprise"] <=5,"anciennete"] = "03 - 05"
df.loc[df["annees_dans_entreprise"] <=2,"anciennete"] = "0 - 02"

# segmentation de l'ancienneté au poste
df["anciennete_poste"] = "21 +"
df.loc[df["annees_poste_actuel"] <=20,"anciennete_poste"] = "10 - 20"
df.loc[df["annees_poste_actuel"] <=10,"anciennete_poste"] = "06 - 10"
df.loc[df["annees_poste_actuel"] <=5,"anciennete_poste"] = "03 - 05"
df.loc[df["annees_poste_actuel"] <=2,"anciennete_poste"] = "0 - 02"

# segmentation du nombre d'années depuis la dernière promotion
df["derniere_promotion"] = "10 +"
df.loc[df["annees_depuis_promotion"] <=10,"derniere_promotion"] = "06 - 10"
df.loc[df["annees_depuis_promotion"] <=5,"derniere_promotion"] = "03 - 05"
df.loc[df["annees_depuis_promotion"] <=2,"derniere_promotion"] = "0 - 02"

# segmentation du nombre d'années avec le manager actuel
df["annees_avec_manager"] = "21 +"
df.loc[df["num_annees_avec_manager"] <=20,"annees_avec_manager"] = "10 - 20"
df.loc[df["num_annees_avec_manager"] <=10,"annees_avec_manager"] = "06 - 10"
df.loc[df["num_annees_avec_manager"] <=5,"annees_avec_manager"] = "03 - 05"
df.loc[df["num_annees_avec_manager"] <=2,"annees_avec_manager"] = "0 - 02"

df
```
